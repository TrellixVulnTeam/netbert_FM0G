## Masked perplexity on dev dataset

Perplexity is just an exponentiation of the entropy: $ùëùùëíùëüùëùùëôùëíùë•ùëñùë°ùë¶=e^{ùëíùëõùë°ùëüùëúùëùùë¶}=e^{loss}$

```python
models = ['BERT-base','NetBERT-300K','NetBERT-800K','NetBERT-1M','NetBERT-1.8M']
#perplexities = [34.6705, 1.4198, 1.3198, 1.3023, 1.2578]
perplexities = [34.671, 1.419, 1.319, 1.302, 1.258]

# Create barplot.
sns.set(font_scale = 2)
plt.figure(figsize=(15,8))
ax = sns.barplot(models, perplexities)

# Rotate labels on x-axis.
ax.set_xticklabels(ax.get_xticklabels(), rotation=0, horizontalalignment='center', fontsize=14)

# Set titles of plot and axes.
plt.ylabel('Masked perplexity', fontsize=18)

#adding the count labels above rectangles.
rects = ax.patches
for rect, label in zip(rects, perplexities):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width()/2, height + 0.2, label, ha='center', va='bottom', fontsize=12)
```
